```python
#import all the needed libraries
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
from scipy import random 

#limit of integration
a = 0
b = 1.75

#number of samples 
N = 200000

#define the function
def func(x): 
    return np.cos(x)

#define the values of cosine from 0 to 1.75
x_of_func = np.linspace(0,1.75,10000)
y_of_func = func(x_of_func)

#define the min and max 
min_of_func = min(y_of_func)
max_of_func = max(y_of_func)

#creating random values
x_random = np.random.uniform(a,b,N)
x_random_2 = np.random.uniform(a,b,N)
y_random = np.random.uniform(min_of_func,max_of_func,N)

#classify where the point is
inside = np.where(func(x_random)/y_random>1)[0]
outside = np.where(func(x_random)/y_random<=1)[0]

#answer of the integral through regular integration (used for graph title, nothing else)
integral_of_cos = float(np.sin(b) - np.sin(a))

#plotting the figure
figure = plt.figure()
plt.grid()
plt.xlim([a,b])
plt.ylim([min_of_func,max_of_func])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("F(0,1.75) = " + str(integral_of_cos))
plt.plot(x_random[inside],y_random[inside],".",color="blue",label="Points below f(x)")
plt.plot(x_random[outside],y_random[outside],".",color="grey",label="Points above f(x)")
plt.plot(x_of_func,y_of_func,color="red",label="f(x)")
plt.legend()

#estimating the value of the integral 
sum_of_x_random = 0.0

for i in range(N):
    sum_of_x_random += func(x_random[i])

value_of_1st_estimate = ((b-a)/float(N))*sum_of_x_random

#get a second estimation 
sum_of_x_random_2 = 0.0

for i in range(N):
    sum_of_x_random_2 += func(x_random_2[i])

value_of_2nd_estimate = ((b-a)/float(N))*sum_of_x_random_2

#tolerance through absolute difference between two numerical estimates of the integral 
tolerance = abs(value_of_1st_estimate - value_of_2nd_estimate)

#printing information
print("Number of samples: ",N)
print("Approximate value of the integral: ",value_of_1st_estimate)
print("Second estimation: ",value_of_2nd_estimate)
print("Tolerance of the approximation: ",tolerance)
```
